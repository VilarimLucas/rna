```python
import numpy as np
import matplotlib.pyplot as plt

# Dados do problema
data = {
    'x1': [0, 0, 1, 1],
    'x2': [0, 1, 0, 1],
    'class': [0, 0, 0, 1]  # Classes C1 = 0 e C2 = 1
}

# Configuração do perceptron
weights = np.array([1, -1])  # Pesos w1 e w2
bias = 0.5  # Bias
activation_function = lambda x: 1 if x >= 0 else 0  # Função degrau

# Função de classificação
def perceptron(x1, x2):
    # Soma ponderada
    weighted_sum = weights[0] * x1 + weights[1] * x2 + bias
    # Aplicar função de ativação
    return activation_function(weighted_sum)

# Testar o perceptron com os dados
classified_data = []
for x1, x2 in zip(data['x1'], data['x2']):
    classified_data.append(perceptron(x1, x2))

# Plotar os padrões de treinamento
plt.figure(figsize=(8, 6))
for x1, x2, c in zip(data['x1'], data['x2'], data['class']):
    color = 'blue' if c == 0 else 'green'
    plt.scatter(x1, x2, color=color, label=f'Classe {c}' if f'Classe {c}' not in plt.gca().get_legend_handles_labels()[1] else "")
plt.axline((0, 0.5), slope=-1, color='red', linestyle='--', label='Fronteira de decisão')
plt.title("Padrões de Treinamento e Fronteira de Decisão")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid()
plt.show()

# Exibir resultados
print("Classificação dos padrões:")
for i, (x1, x2, c) in enumerate(zip(data['x1'], data['x2'], classified_data)):
    print(f"Padrão {i+1}: (x1={x1}, x2={x2}) -> Classe prevista: {c}")
```
